# Part A – Application area review.

AI has been widely applied in the problem domain of e-mail and spam filtering, with a focus on using machine learning algorithms to automatically classify e-mails as spam or not. The problem of spam filtering is a classic text classification problem. Spam filtering is the process of identifying unwanted e-mails and separating them from legitimate e-mails. The goal of spam filtering is to reduce the amount of unwanted e-mails that users receive while allowing important e-mails to pass through.

One of the most commonly used techniques in this domain is the use of Bayesian filters, which use Bayes' theorem to calculate the probability that an e-mail is spam based on the presence of certain words or phrases. These filters can be trained on a dataset of labeled e-mails, and have been shown to be highly effective in detecting spam. Bayesian filters have the advantage of being able to adapt to new spam as it appears, and they can also be easily integrated into existing e-mail systems.

Another popular technique is the use of support vector machines (SVMs) to classify e-mails. SVMs are a type of supervised learning algorithm that can be used to find the best boundary between different classes of data. They have been used in e-mail filtering to identify the key features that distinguish spam e-mails from legitimate ones, and have been shown to have a high accuracy rate in detecting spam. SVMs have the advantage of being able to handle high-dimensional data and they can also be used for non-linear classification.

Another notable technique is the use of Artificial Neural Networks (ANNs) for email filtering. ANNs are a type of machine learning algorithm inspired by the structure and function of the human brain. They have been used in e-mail filtering to identify patterns and characteristics in e-mails that are indicative of spam. ANNs can be trained on a dataset of labeled e-mails and have been shown to be highly effective in detecting spam. ANNs can also be used for feature selection and feature extraction which can improve the overall performance of the spam filtering system.

Another technique is the use of Natural Language Processing (NLP) for e-mail filtering. NLP is a subfield of AI that deals with the analysis and manipulation of human language. NLP techniques have been used in e-mail filtering to extract semantic information from e-mails, such as the meaning of the words and phrases used, and to identify patterns and characteristics that are indicative of spam. NLP techniques have been used to extract features such as the subject, body, sender, recipient, and the presence of certain words and phrases, which can be used to train a spam classifier.

Recent research has also been focused on the use of deep learning models, such as recurrent neural networks (RNNs) and convolutional neural networks (CNNs) for e-mail filtering. These models have shown to be very effective in detecting spam, with accuracy rates of over 99%. The advantage of deep learning models is their ability to automatically extract features from the raw data, which can improve the overall performance of the spam filtering system.

Hybrid models have also been proposed in the literature, which combine the advantages of multiple techniques to improve the performance of the spam filtering system. For example, a combination of a Bayesian filter and an SVM has been proposed, which can improve the overall performance of the spam filtering system by combining the advantages of both techniques.

Overall, AI has been successfully applied to the problem domain of e-mail and spam filtering, with a variety of techniques showing high accuracy rates in detecting spam. Bayesian filters, SVMs, ANNs, NLP, deep learning models and Hybrid models have been the most relevant techniques used



# Part B – Compare and evaluate AI techniques

## Support Vector Machine (SVM)

The goal of using SVM in the problem domain of e-mail and spam filtering is to identify the key features that distinguish spam e-mails from legitimate ones, and to use these features to classify new e-mails as spam or not.

SVMs are a type of supervised learning algorithm that can be used to find the best boundary between different classes of data. They can be used to classify e-mails by learning from a dataset of labeled e-mails, where the features are extracted from the e-mails, such as the presence of certain words or phrases, and the label is whether the e-mail is spam or not. The SVM algorithm uses a linear boundary to separate spam and non-spam e-mails. The algorithm finds the best boundary by maximizing the distance between the closest points from each class, which are called support vectors. One of the main advantages of using SVMs for email and spam filtering is their ability to handle high-dimensional data, such as text data. SVMs use a technique called kernel trick to transform the input data into a higher dimensional space, where it is easier to find a linear boundary between different classes. Additionally, SVMs are robust to overfitting and have a regularization parameter that can be used to control the complexity of the model. One of the disadvantages of SVMs is that they can be sensitive to the choice of kernel function and the parameters of the kernel. If the kernel function is not chosen properly, the performance of the SVM can suffer. Additionally, SVMs can be computationally expensive for large datasets and can take a long time to train.

The strengths of using SVM for e-mail filtering are:

1. They can handle high-dimensional data, which is useful when working with text data, where the number of possible features can be very large.
2. They can also be used for non-linear classification, which can improve the accuracy of the classifier when the data is not linearly separable.
3. SVMs have been widely used in the literature and have shown to have a high accuracy rate in detecting spam.
4. SVMs are not sensitive to the presence of irrelevant features.

The weaknesses of using SVM for e-mail filtering are:

1. The training time for SVMs can be relatively slow, especially when working with large datasets.
2. SVMs can be sensitive to the choice of kernel function, which can affect the performance of the classifier.
3. SVMs can be sensitive to the presence of outliers.
4. SVMs can be sensitive to the presence of noise in the data.

## Bayesian filters

The goal of using Bayesian filters in the problem domain of e-mail and spam filtering is to calculate the probability that an e-mail is spam based on the presence of certain words or phrases, and to use this probability to classify new e-mails as spam or not.

Bayesian filters are based on Bayes' theorem, which states that the probability of an event occurring given that another event has occurred is equal to the probability of the two events occurring together divided by the probability of the second event occurring. Bayesian filters can be trained on a dataset of labeled e-mails, where the features are the presence of certain words or phrases, and the label is whether the e-mail is spam or not. The algorithm uses these probabilities to calculate the probability that a new e-mail is spam, and classifies the e-mail as spam or not based on a threshold probability. One of the main advantages of using Bayesian filters is that they are computationally simple and can be easily implemented. Additionally, Bayesian filters can be trained on the fly, allowing them to adapt to new types of spam as they appear. One of the disadvantages of Bayesian filters is that they can be prone to errors if the training data is not representative of the actual emails being filtered. Additionally, Bayesian filters can have a high rate of false positives, incorrectly marking legitimate emails as spam.

The strengths of using Bayesian filters for e-mail filtering are:

1. They can be easily integrated into existing e-mail systems.
2. They can adapt to new spam as it appears.
3. Bayesian filters have been widely used in the literature and have shown to have a high accuracy rate in detecting spam.

The weaknesses of using Bayesian filters for e-mail filtering are:

1. They can be sensitive to the presence of irrelevant features.
2. They can be sensitive to the presence of noise in the data.
3. They can be sensitive to the presence of outliers.
4. Bayesian filters can be sensitive to the presence of new words or phrases that have not been seen before, which can affect the performance of the classifier.

In terms of input data, SVM and ANN algorithms require numerical data, and a labeled dataset to train the model. Bayesian filters require labeled dataset but it is not necessary to be numerical. As for the expected output, all the three techniques, SVM, ANN, and Bayesian filters will output a probability value of the email being spam or not.

The Author of this project has decided to move forward with the SVM technique. As SVM can be a better choice over Bayesian Filters and ANN for e-mail and spam filtering because:

1. SVMs can handle high-dimensional data, which is useful when working with text data, where the number of possible features can be very large.
2. SVMs can also be used for non-linear classification, which can improve the accuracy of the classifier when the data is not linearly separable.
3. SVM is robust to overfitting, it has regularization parameter "C" to control the trade-off between maximizing the margin and minimizing the classification error.
4. SVMs are not sensitive to the presence of irrelevant features which is common in the email classification task.
5. SVMs have been widely used in the literature and have shown to have a high accuracy rate in detecting spam.

Support Vector Machines (SVMs) are a popular technique for text classification tasks such as email and spam filtering. When using SVMs for this task, appropriate evaluation metrics would include precision, recall, accuracy, and F1 score.

Accuracy measures the proportion of correctly classified emails (both spam and non-spam) out of all emails. F1 score is a balance between precision and recall, and can be useful for comparing different models or model configurations.

Importing the Dependencies

Label Encoding

```python
# label spam mail as 0;  non-spam/ham mail as 1;

mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
# separating the data as texts and label

X = mail_data['Message']

Y = mail_data['Category']
```

Splitting the data into training data & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to the SVM

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

Training the Model

## SVM

```python
model = svm.SVC(kernel='rbf')
```

```python
# training the SVM model with the training data
model.fit(X_train_features, Y_train)
```

Evaluating the trained model

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

**Precision, Recall, & F1 Score - function**

```python
def precision_recall_f1_score(true_labels, pred_labels):

  precision_value = precision_score(true_labels, pred_labels)
  recall_value = recall_score(true_labels, pred_labels)
  f1_score_value = f1_score(true_labels, pred_labels)

  print('Precision =',precision_value)
  print('Recall =',recall_value)
  print('F1 Score =',f1_score_value)
```

```python
# classification metrics for training data
precision_recall_f1_score(Y_train, prediction_on_training_data)
```

```python
# classification metrics for test data
precision_recall_f1_score(Y_test, prediction_on_test_data)
```

Benchmark Analysis

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier
```

```python
knc = KNeighborsClassifier()
mnb = MultinomialNB()
bc = BaggingClassifier(n_estimators=50, random_state=2)
lgr = LogisticRegression()
```

```python
clfs = {
    'KN' : knc, 
    'NB': mnb, 
    'LR': lgr,
    'SVM': model,
    'BgC': bc
}
```

```python
def train_classifier(clf,X_train,y_train,X_test,y_test):
    clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    
    return accuracy,precision, f1, recall
```

```python
accuracy_scores = []
precision_scores = []
f1_scores = []
recall_scores = []

for name,clf in clfs.items():
    
    current_accuracy,current_precision, current_f1, current_recall = train_classifier(clf, X_train_features,Y_train,X_test_features,Y_test)
    
    print("For ",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)
    print("F1 Score - ",current_f1)
    print("Recall - ",current_recall)
    
    accuracy_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    f1_scores.append(current_f1)
    recall_scores.append(current_recall)
```

```python
performance_df = pd.DataFrame({'Algorithm':clfs.keys(),'Accuracy':accuracy_scores,'Precision':precision_scores, 'F1 Score': f1_scores, 'Recall': recall_scores}).sort_values('Accuracy',ascending=False)
```
